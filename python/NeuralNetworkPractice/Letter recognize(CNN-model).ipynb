```python
import numpy as np
import csv
from sklearn.model_selection import train_test_split
import pandas
from PIL import Image
from tensorflow.keras import layers,models
```

```python
def file_set():
    #used english handwright dataset from
    #https://www.kaggle.com/datasets/dhruvildave/english-handwritten-characters-dataset?resource=download
    #It contains 62 classes with 55 images of each class. The 62 classes are 0-9, A-Z and a-z.
    file_name = 'english.csv'
    data_csv=pandas.read_csv(file_name)
    data_list=pandas.DataFrame(data_csv)
    
    image_link=data_list['image']
    target=data_list['label']
    
    x_train_,x_test_,y_train_,y_test_=train_test_split(image_link,target,test_size=0.3,random_state=41)
    
    #make target value from string to interger to make as the same as output node index
    y_train,y_test=[],[]
    for k_te in y_test_:
        if 'A'<=k_te<='Z':
            y_test.append(ord(k_te)-ord('A')+10)
        elif 'a'<=k_te<='z':
            y_test.append(ord(k_te)-ord('a')+36)
        else:
            y_test.append(int(k_te))
    for k_tr in y_train_:
        if 'A'<=k_tr<='Z':
            y_train.append(ord(k_tr)-ord('A')+10)
        elif 'a'<=k_tr<='z':
            y_train.append(ord(k_tr)-ord('a')+36)
        else:
            y_train.append(int(k_tr))

    #make image data as interger array
    x_train,x_test=[],[]
    for key in x_train_:
        im = Image.open(key)
        im_data=abs(np.asfarray(np.array(im.getdata()))-255)#to make value from 0-1
        im_data=np.split(im_data,27)#reformat array for tensorflow module
        x_train.append(im_data)
    for key in x_test_:
        im = Image.open(key)
        im_data=abs(np.asfarray(np.array(im.getdata()))-255)
        im_data=np.split(im_data,27)
        x_test.append(im_data)
    
    return np.array(x_train),np.array(x_test),np.array(y_train),np.array(y_test)
```

```python
def ModelSet():
    #data setting
    x_train,x_test,y_train,y_test=file_set()
    
    #model setting
    model=models.Sequential()
    
    #input and hidden node layers
    model.add(layers.Conv2D(62, (3, 3), activation='relu', input_shape=(27, 27, 1)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(62, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(62, (3, 3), activation='relu'))
    
    #final output layers
    model.add(layers.Flatten())
    model.add(layers.Dense(62, activation='relu'))
    model.add(layers.Dense(62, activation='softmax'))
    
    #set
    model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=50)#train
    
    return model
```

```python
model=ModelSet()
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print('\nperformance:',test_acc)
```
